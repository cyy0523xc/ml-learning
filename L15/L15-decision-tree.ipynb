# 决策树

## 1. 决策树生成图

```python
from sklearn import tree
from sklearn.datasets import load_iris
import pydotplus

# 载入sklearn中自带的数据Iris，构造决策树

iris = load_iris()
# clf = tree.DecisionTreeClassifier(max_depth=3, criterion='entropy')
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(iris.data, iris.target)

from IPython.display import Image
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

## 2. 使用sklearn进行分类

```python
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score

# 导入数据并处理成二分类
iris = load_iris()
X = iris.data
y = iris.target

# shuffle and split training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)

# model = tree.DecisionTreeClassifier(max_depth=3, criterion='entropy')
model = tree.DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)

y_pred = model.predict(X_train)
print("训练误差：")
print(precision_score(y_train, y_pred, average='macro'))
print(recall_score(y_train, y_pred, average='macro'))
print(f1_score(y_train, y_pred, average='macro'))

y_pred = model.predict(X_test)
print("测试误差：")
print(precision_score(y_test, y_pred, average='macro'))
print(recall_score(y_test, y_pred, average='macro'))
print(f1_score(y_test, y_pred, average='macro'))
```
